```python
# example


# Core Library modules
import os
import random as rn

# Third party modules
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import tensorflow as tf
#from tensorflow.keras.utils import print_summary
```

```python
# First party modules
#import N2D
from N2D import AutoEncoder, UmapGMM, load_n2d, n2d, save_n2d
```

```python
os.environ["PYTHONHASHSEED"] = "0"
os.environ["TF_CUDNN_USE_AUTOTUNE"] = "0"

rn.seed(0)
tf.random.set_seed(0)


np.random.seed(0)
```

```python
from datasets import load_mnist, load_mnist_test, load_usps, load_pendigits, load_fashion, load_har
x,y=load_mnist()
n_clusters = 10
print(x.shape[-1])
```

```python
ae = AutoEncoder(x.shape[-1], n_clusters)
```

```python
manifoldGMM = UmapGMM(n_clusters)
mnistcluster = n2d(ae, manifoldGMM )
```

```python
mnistcluster.fit(x, weight_id = "mnist.h5",patience=20)
```

```python
preds = mnistcluster.predict(x)
mnistcluster.visualize(y, None, n_clusters=n_clusters)
plt.show()
```

```python
print(mnistcluster.assess(y))
```
